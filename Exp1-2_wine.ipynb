필요한 모듈 import

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score # 정확도 평가하기

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
```

Data 준비

```python
# Data load
wine = load_wine()

# Feature Data 지정
wine_data = wine.data

# Label Data 지정
wine_label = wine.target
print(wine_label.shape)

# Target Names 출력해보기
print(wine.target_names)

# Data Describe 해보기
print(wine.DESCR)
```

Train, Test Data 분리

```python
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
```

다양한 모델로 학습 시켜보기

```python
# 모델을 입력으로 받아서 정확도를 report해주는 함수
def accur(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("This is " + str(model))
    print(f'Accurancy: {accuracy}')
    print(classification_report(y_test, y_pred))
    print('---------------------------------------')
    print()

# Decision Tree
accur(DecisionTreeClassifier(random_state=32))


# Random Forest
accur(RandomForestClassifier(random_state=32))

# SVM
accur(svm.SVC())

# SGD Classifier
accur(SGDClassifier())

# Logistic Regression
accur(LogisticRegression())
```
